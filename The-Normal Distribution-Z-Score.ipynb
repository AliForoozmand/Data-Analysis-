## The Normal Distribution - Z Score

### The Z score is a standardization technique commonly employed for nearly normal observations but that may be used with any distribution. It is defined as the number of standard deviations that an observed event falls above of below the mean. It is expressed mathematically as: 

$$Z = \frac{x-\mu}{\sigma}$$

### By using a *normal probability table* which lists Z scores and percentiles, we can identify a percentile based on the Z score. The percentile represents the probability of having an event with a lower Z score.

### Question 1: Suppose that the developing time, *X*, for a particular Kodak photographic printing paper after exposure to a 5 second light source in the darkroom is normally distributed with a mean of 25 seconds and a standard deviation of 1.3 seconds. Determine the probability of needing 25 seconds or more to develop the film, that is $P(X \geq 25)$.

```python
import random
import matplotlib.pyplot as plt
import numpy as np 


def z (x, mu, sig):
    return (x-mu) / sig

print (f" Z = {z(25,25, 1.3)}")
print ("for Z = 0, Regarding table we have 0.5 probability")
print ("so for P(X>=25) = 50% ")
```

### Question 2: Determine the probability that a particular film will take more than 26.5 seconds to develop.

```python
print (f"Z = {z(26.5, 25, 1.3)}")
print (f"for Z = {z(26.5, 25, 1.3)}, Regarding table we have 0.87493 probability, that this is probabilty less than 26.5 so for more than 26.5 we should do this : 1- P(X< 26.5)")
print (f"so The probability of P(x>=26.5) we have {1-0.87493}")
print ("so for P(X>=26.5) = 12.5% ")
```

### Question 3: What is the probability that the developing time is less than 23 seconds?

```python
print (f"Z = {z(23, 25, 1.3)}")
print (f"for Z = {z(23, 25, 1.3)}, Regarding table we have 0.06178 probability")
print ("so for P(X<23) = 6.2% ")
```

# we also can solve these questions on the other way:

```python
from scipy.stats import norm

mu = 25 
sigma = 1.3  

# Question 1: P(X >= 25)
# For a normal distribution, P(X >= mu) is always 0.5.
prob_25_or_more = 1 - norm.cdf((25 - mu) / sigma)

# Question 2: P(X > 26.5)
z_26_5 = (26.5 - mu) / sigma
prob_more_than_26_5 = 1 - norm.cdf(z_26_5)

# Question 3: P(X < 23)
z_23 = (23 - mu) / sigma
prob_less_than_23 = norm.cdf(z_23)

# Question 4: P(|X - mu| > 1.5)
# This is equivalent to finding P(X < 23.5 or X > 26.5)
z_23_5 = (23.5 - mu) / sigma
z_26_5_diff = (26.5 - mu) / sigma
prob_diff_more_than_1_5 = norm.cdf(z_23_5) + (1 - norm.cdf(z_26_5_diff))

print (f"question 1 : {prob_25_or_more * 100} %")
print (f"question 2 : {prob_more_than_26_5* 100} %")
print (f"question 3 : {prob_less_than_23* 100} %")
print (f"question 4 : {prob_diff_more_than_1_5* 100} %")
```

## Given random variable $X_1$, with PDF $f(x) = exp(-x)$ in the interval [0,Z]. Define 3 new random variables $X_2$, $X_4$ and $X_8$ which are the sum of 2, 4 and 8 random variables $X_1$, respectively.

### Question 5: Do a numerical simulation by generating N = 10000 random numbers for each of the four $X_i$. Plot the four histograms.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math 
np.random.seed(0)

N = 10000  
num_bins = 50  

X1 = np.random.exponential(scale=1, size=N)        
X2 = np.random.exponential(scale=1, size=(N, 2)).sum(axis=1)  
X4 = np.random.exponential(scale=1, size=(N, 4)).sum(axis=1)  
X8 = np.random.exponential(scale=1, size=(N, 8)).sum(axis=1)  
    
plt.figure(figsize=(12, 8))

# X1
plt.subplot(2, 2, 1)
plt.hist(X1, bins=num_bins, density= True, edgecolor = "black")
plt.title("$X_1$ (1 exp sample)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# X2
plt.subplot(2, 2, 2)
plt.hist(X2, bins=num_bins, density= True,color='teal', edgecolor='black')
plt.title("$X_2$ (sum of 2 exp samples)")
plt.xlabel("Value")

# X4
plt.subplot(2, 2, 3)
plt.hist(X4, bins=num_bins,density= True, color='lightgreen',edgecolor='black')
plt.title("$X_4$ (sum of 4 exp samples)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# X8
plt.subplot(2, 2, 4)
plt.hist(X8, bins=num_bins,density= True, color='coral',edgecolor='black')
plt.title("$X_8$ (sum of 8 exp samples)")
plt.xlabel("Value")

plt.tight_layout()
plt.show()
```

### Question 6: Calculate mean value, variance, skewness and excess kurtosis for each of the four $X_i$ numerically using the generated data. Plot them as a function of the number $n$ of added random numbers (each of the four plots consists of four data points at $x$ coordinates 1, 2, 4 and 8). These functions are all governed by power laws $n^k$. For each case estimate the exponent $k$ (e.g. by looking at a log-log plot).

```python
from scipy.stats import skew, kurtosis

N = 10000
num_bins = 50

sum_counts = [1, 2, 4, 8]

mean_values = []
variance_values = []
skewness_values = []
kurtosis_values = []

for n in sum_counts:
    X_n = np.random.exponential(scale=1, size=(N, n)).sum(axis=1)
    
    mean_values.append(np.mean(X_n))
    variance_values.append(np.var(X_n))
    skewness_values.append(skew(X_n))
    kurtosis_values.append(kurtosis(X_n))

plt.figure(figsize=(12, 10))

# Plot Mean
plt.subplot(2, 2, 1)
plt.plot(sum_counts, mean_values, 'o-', color='blue')
plt.xscale("log")
plt.yscale("log")
plt.ylabel("Mean")
plt.title("Mean")
for i, txt in enumerate(mean_values):
    plt.annotate(f'{txt:.2f}', (sum_counts[i], mean_values[i]), textcoords="offset points", xytext=(5,5), ha='center')

    
    
# Plot Variance
plt.subplot(2, 2, 2)
plt.plot(sum_counts, variance_values, 'o-', color='orange')
plt.xscale("log")
plt.yscale("log")
plt.ylabel("Variance")
plt.title("Variance")
for i, txt in enumerate(variance_values):
    plt.annotate(f'{txt:.2f}', (sum_counts[i], variance_values[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Plot Skewness
plt.subplot(2, 2, 3)
plt.plot(sum_counts, skewness_values, 'o-', color='green')
plt.xscale("log")
plt.yscale("log")
plt.ylabel("Skewness")
plt.title("Skewness ")
for i, txt in enumerate(skewness_values):
    plt.annotate(f'{txt:.2f}', (sum_counts[i], skewness_values[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Plot Excess Kurtosis
plt.subplot(2, 2, 4)
plt.plot(sum_counts, kurtosis_values, 'o-', color='red')
plt.xscale("log")
plt.yscale("log")
plt.ylabel("Excess Kurtosis")
plt.title("Excess Kurtosis")
for i, txt in enumerate(kurtosis_values):
    plt.annotate(f'{txt:.2f}', (sum_counts[i], kurtosis_values[i]), textcoords="offset points", xytext=(5,5), ha='center')

plt.tight_layout()
plt.show()
```

As we can see in plots, The plot of Mean and Variance governed by power low $n^k$ roughly, which are $n^0$, $n^1$, $n^2$, $n^3$,  but in the Skewness and The Kurtosis you can't see anything. 

### Question 7: Analytically calculate the PDF for $X_2$ by using the convolution integral (see lecture notes; there the notation $Z = X + Y$ was used). Overlay it to the corresponding histogram (scale the function as necessary).

### Final PDF:

$f_{X_2}(z) = z e^{-z},\quad z\geq 0.$


```python
N = 10000
num_bins = 50

X2 = np.random.exponential(scale=1, size=(N, 2)).sum(axis=1)

def pdf_X2(z):
    return z * np.exp(-z) if z >= 0 else 0

z_values = np.linspace(0, max(X2), 1000)
pdf_values = np.array([pdf_X2(z) for z in z_values])


plt.figure(figsize=(8, 6))
plt.hist(X2, bins=num_bins, density=True, edgecolor='black',color ="teal", label='Histogram of $X_2$')
plt.plot(z_values, pdf_values, 'r-', label='Analytical PDF ($z e^{-z}$)', linewidth=2)
plt.title('PDF of $X_2 = X_1 + X_1$')
plt.xlabel('$z$')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
