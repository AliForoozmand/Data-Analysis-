### Question 2: Using the package Matplotlib, create a 10-point random data plot and a 1000 random points plot. Customize the axes, change the lines to points. Present both plots in the same figure.

### Question 9: Suppose that I have only a 10% change of hitting the bullseye in any attempt during a game of darts. Determine the probability that I hit the bullseye at least two times in a game consisting of 10 attempts. Use your combination formula and the binomial distribution to calculate this probability in Python.

```python
from scipy.stats import binom
n = 10 
p = .1

# for the probability that hit the bullseye at least two times we should calculate in this way
p1 = binom.pmf(k=1, n=n, p=p) 
p0 = binom.pmf(k=0, n=n, p=p)

total_prob_2times = 1 - (p1 + p0)

print (total_prob_2times)
```
